# Function Application and Mapping

```python
 frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
         index=['Utah', 'Ohio', 'Texas', 'Oregon'])
```

```python
frame
```

```python
np.abs(frame)          # displays value in positive
```

```python
frame.max()  # by default axis is zero means sure it is extracting values vertically but 
             # to do so it is accessing every row element
```

```python
frame.min()
```

```python
f = lambda x: x.max() - x.min()
frame.apply(f)                   # pooray frame par koi function implement karna
```

```python
f = lambda x: x**2
frame.apply(f)
```

```python
frame
```

```python
MAX = frame.max(axis = 1)
MAX
```

```python
MIN = frame.min(axis = 1)
MIN
```

```python
f = lambda x: MAX - MIN
frame.apply(f)                   # pooray frame par koi function implement karna
```

# Correlation and Covariance

```python
import pandas_datareader.data as web
```

```python
# dictionary comprehension

all_data = {ticker: web.get_data_yahoo(ticker)  # key(ticker) is not a string but a variable means it can change,
                                                # similarly, value is a variable as well
 for ticker in ['AAPL', 'IBM', 'MSFT', 'GOOG']}
```

```python
all_data
```

```python
#data = pd.DataFrame(all_data)
```

```python
price = pd.DataFrame({ticker: data['Adj Close']  #taking one column, ticker tou saaray liye hain
 for ticker, data in all_data.items()})
```

```python
price
```

```python
volume = pd.DataFrame({ticker: data['Volume']
 for ticker, data in all_data.items()})
volume
```

```python
returns = price.pct_change()
```

```python
returns.tail()
```

```python
returns['MSFT'].corr(returns['IBM'])
```

```python
returns['MSFT'].cov(returns['IBM'])
```

```python
returns.corr()
```

```python
returns.cov()
```
